```python
# Importing header files
import numpy as np
import warnings

warnings.filterwarnings('ignore')

#New record
new_record=[[50,  9,  4,  1,  0,  0, 40,  0]]

#Reading file
data = np.genfromtxt("census.csv", delimiter=",", skip_header=1)

census = np.concatenate((new_record,data))

#Code starts here
age  = census[:,0]
max_age = age.max()
min_age = age.min()
age_mean = age.mean()
age_mean = age_mean.round(2)
age_std = age.std()
age_std = age_std.round(2)



race = census[:,2]

race_0 ,race_1 , race_2 ,race_3 , race_4= [] ,[] , [], [] ,[]

for races in race:
    if races == 0 :
        race_0 = np.append(race_0,[races])
    if races == 1 :
        race_1 = np.append(race_1,[races])
    if races == 2 :
        race_2 = np.append(race_2,[races])
    if races == 3 :
        race_3 = np.append(race_3,[races])
    if races == 4 :
        race_4 = np.append(race_4,[races])

len_0 = len(race_0)
len_1 = len(race_1)
len_2 = len(race_2)
len_3 = len(race_3)
len_4 = len(race_4)


min_race = min(len_0,len_1,len_2,len_3,len_4)

if len_0 == min_race :
    minority_race = 0
if len_1 == min_race :
    minority_race = 1
if len_2 == min_race :
    minority_race = 2
if len_3 == min_race :
    minority_race = 3
if len_4 == min_race :
    minority_race =4

senior_citizens = census[:,0]
working_hours = census[:,6]

working_hours_sum = np.sum(working_hours)

avg_working_hours = working_hours_sum / len(senior_citizens)
avg_working_hours = avg_working_hours.round(2)

high = census[census[:,1]> 10]

low = census[census[:,1]<=10]

high_sum = high[:,7]
low_sum = low[:,7]
avg_pay_high = np.mean(high_sum)

avg_pay_low = np.mean(low_sum)

print(avg_pay_high.round(2))
print(avg_pay_low.round(2))











```
